```python
import matplotlib.image as mpimg
import cv2 
from google.colab.patches import cv2_imshow
# Use this For adding noise
filename = '/content/drive/MyDrive/Tuberculosis/Tuberculosis-1230.png'

im = mpimg.imread(filename)
plt.figure(figsize=(15,12))
sigmas = [ .12, .25, .35 ]
for i in range(1):
 noisy = cv2.cvtColor(im, cv2.COLOR_GRAY2RGB)
 noisy = random_noise(noisy, var=sigmas[i]**2)
 ##plt.subplot(2,2,i+1)
 plt.imshow(noisy)
 plt.axis('off')
 plt.title('Gaussian noise with sigma=' + str(sigmas[i]), size=20)
 print(noisy.shape)
plt.tight_layout()
plt.show()
```
