```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import os
```

```python
img = cv2.imread("/content/drive/MyDrive/Gaussian/Normal/Normal-2.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(11,9))
print(img.shape)
plt.imshow(img,cmap='gray')
plt.show()
```

```python
hist=cv2.calcHist(img,[0],None,[256],[0,256])
plt.plot(hist)
plt.show()
```

```python
img_equalize = cv2.equalizeHist(img)
plt.figure(figsize=(11,9))
plt.subplot(121), plt.title("Original"), plt.imshow(img,cmap='gray')
plt.subplot(122), plt.title("equalized contrast Image"), plt.imshow(img_equalize,cmap='gray')
plt.show()
```

# Contrast Limited Adaptive Histogram Equalization

```python
clahe = cv2.createCLAHE(clipLimit=5)
clahe_img = clahe.apply(img_equalize)

plt.figure(figsize=(20,13))
plt.subplot(131), plt.title("Original"), plt.imshow(img,cmap='gray')
plt.subplot(132), plt.title("equalized contrast Image"), plt.imshow(img_equalize,cmap='gray')
plt.subplot(133), plt.title("clahe Image"), plt.imshow(clahe_img,cmap='gray')
plt.show()
```

# Thresholding 

```python
threshold = 120
max_value = 255
```

### Binary Thresholding

```python
ret1, i1 = cv2.threshold(clahe_img, threshold, max_value, cv2.THRESH_BINARY)

plt.figure(figsize=(20,13))
plt.subplot(131), plt.title("Original"), plt.imshow(img,cmap='gray')
plt.subplot(132), plt.title("clahe Image"), plt.imshow(clahe_img,cmap='gray')
plt.subplot(133), plt.title("Binary Threshold"), plt.imshow(i1,cmap='gray')
plt.show()
```

#### Trunc Threshold

```python
ret2, i2 = cv2.threshold(clahe_img, threshold, max_value, cv2.THRESH_TRUNC)

plt.figure(figsize=(20,13))
plt.subplot(131), plt.title("Original"), plt.imshow(img,cmap='gray')
plt.subplot(132), plt.title("clahe Image"), plt.imshow(clahe_img,cmap='gray')
plt.subplot(133), plt.title("Trunc Threshold"), plt.imshow(i2,cmap='gray')
plt.show()
```
